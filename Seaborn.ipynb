# 🟢 BASIC PLOTS

---

## 1. 📊 Distribution Plot — histplot

> Shows how data is distributed!

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

# Basic histogram
sns.histplot(data=tips, x='total_bill')
plt.title('Distribution of Total Bill')
plt.show()

# With KDE curve (smooth line)
sns.histplot(data=tips, x='total_bill', kde=True)
plt.show()

# Colored by category
sns.histplot(data=tips, x='total_bill',
             hue='sex', kde=True)
plt.show()

# Change number of bins
sns.histplot(data=tips, x='total_bill', bins=30)
plt.show()
```

---

## 2. 📈 KDE Plot — kdeplot

> Smooth version of histogram — shows probability density!

```python
# Basic KDE
sns.kdeplot(data=tips, x='total_bill')
plt.show()

# Filled KDE
sns.kdeplot(data=tips, x='total_bill', fill=True)
plt.show()

# Multiple categories
sns.kdeplot(data=tips, x='total_bill',
            hue='sex', fill=True)
plt.show()

# 2D KDE — two variables
sns.kdeplot(data=tips, x='total_bill', y='tip')
plt.show()
```

---

## 3. 📦 Box Plot — boxplot

> Shows median, quartiles and outliers!

```python
# Basic box plot
sns.boxplot(data=tips, x='day', y='total_bill')
plt.show()

# With color grouping
sns.boxplot(data=tips, x='day', y='total_bill',
            hue='sex')
plt.show()

# Horizontal box plot
sns.boxplot(data=tips, x='total_bill', y='day',
            orient='h')
plt.show()

# What box plot shows:
# ─────────────────
# Top whisker    → max (excluding outliers)
# Top box edge   → Q3 (75th percentile)
# Middle line    → Median (50th percentile)
# Bottom box edge→ Q1 (25th percentile)
# Bottom whisker → min (excluding outliers)
# Dots           → outliers!
```

---

## 4. 🎻 Violin Plot — violinplot

> Combines box plot with KDE — shows distribution shape!

```python
# Basic violin
sns.violinplot(data=tips, x='day', y='total_bill')
plt.show()

# With hue
sns.violinplot(data=tips, x='day', y='total_bill',
               hue='sex')
plt.show()

# Split violin
sns.violinplot(data=tips, x='day', y='total_bill',
               hue='sex', split=True)
plt.show()
```

---

## 5. 📊 Bar Plot — barplot

> Shows mean of a variable with confidence interval!

```python
# Basic bar plot
sns.barplot(data=tips, x='day', y='total_bill')
plt.show()

# With hue
sns.barplot(data=tips, x='day', y='total_bill',
            hue='sex')
plt.show()

# Change estimator
import numpy as np
sns.barplot(data=tips, x='day', y='total_bill',
            estimator=np.median)  # show median instead!
plt.show()
```

---

## 6. 📊 Count Plot — countplot

> Counts occurrences of each category!

```python
# Basic count
sns.countplot(data=tips, x='day')
plt.show()

# With hue
sns.countplot(data=tips, x='day', hue='sex')
plt.show()

# Horizontal
sns.countplot(data=tips, y='day')
plt.show()

# Order by count
order = tips['day'].value_counts().index
sns.countplot(data=tips, x='day', order=order)
plt.show()
```

---

## 7. 🔵 Scatter Plot — scatterplot

> Shows relationship between two numeric variables!

```python
# Basic scatter
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()

# With color grouping
sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='sex')
plt.show()

# With size grouping
sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='sex', size='size')
plt.show()

# With style grouping
sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='sex', style='time')
plt.show()
```

---

## 8. 📈 Line Plot — lineplot

> Shows trends over time or ordered data!

```python
flights = sns.load_dataset('flights')

# Basic line plot
sns.lineplot(data=flights, x='year', y='passengers')
plt.show()

# With confidence interval
sns.lineplot(data=flights, x='year', y='passengers',
             hue='month')
plt.show()

# Custom line style
sns.lineplot(data=flights, x='year', y='passengers',
             linewidth=2, color='red', marker='o')
plt.show()
```

## 9. 🔥 Heatmap — heatmap

> Most important for ML — shows correlation matrix!

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation heatmap — most used in ML!
tips = sns.load_dataset('tips')
corr = tips.corr(numeric_only=True)

sns.heatmap(corr)
plt.show()

# With annotations and colors
sns.heatmap(corr,
            annot=True,       # show numbers
            fmt='.2f',        # 2 decimal places
            cmap='coolwarm',  # color scheme
            linewidths=0.5)   # lines between cells
plt.title('Correlation Heatmap')
plt.show()

# Other color maps:
# 'Blues', 'Reds', 'Greens'
# 'YlOrRd', 'viridis', 'plasma'
# 'RdYlGn', 'coolwarm', 'seismic'
```

---

## 10. 🔵 Pair Plot — pairplot

> Shows all combinations of variables at once!

```python
iris = sns.load_dataset('iris')

# Basic pair plot
sns.pairplot(iris)
plt.show()

# With color grouping
sns.pairplot(iris, hue='species')
plt.show()

# Customize diagonal plots
sns.pairplot(iris, hue='species',
             diag_kind='kde')  # kde or hist
plt.show()
```

---

## 11. 🎯 Strip Plot — stripplot

> Shows all individual data points!

```python
# Basic strip plot
sns.stripplot(data=tips, x='day', y='total_bill')
plt.show()

# With jitter to avoid overlap
sns.stripplot(data=tips, x='day', y='total_bill',
              jitter=True, hue='sex')
plt.show()
```

---

## 12. 🎯 Swarm Plot — swarmplot

> Like strip plot but points never overlap!

```python
# Swarm plot
sns.swarmplot(data=tips, x='day', y='total_bill')
plt.show()

# Combined with box plot!
sns.boxplot(data=tips, x='day', y='total_bill',
            color='lightblue')
sns.swarmplot(data=tips, x='day', y='total_bill',
              color='black', size=3)
plt.show()
```

---

## 13. 📊 ecdfplot

> Shows cumulative distribution!

```python
sns.ecdfplot(data=tips, x='total_bill')
plt.show()

sns.ecdfplot(data=tips, x='total_bill', hue='sex')
plt.show()
```

## 14. 🎨 FacetGrid — Multiple plots in grid!

> Creates a grid of plots for different categories!

```python
tips = sns.load_dataset('tips')

# Create grid
g = sns.FacetGrid(tips, col='time', row='sex')
g.map(sns.histplot, 'total_bill')
plt.show()

# Scatter in grid
g = sns.FacetGrid(tips, col='day', hue='sex')
g.map(sns.scatterplot, 'total_bill', 'tip')
g.add_legend()
plt.show()
```

---

## 15. 🎨 Joint Plot — jointplot

> Shows relationship + individual distributions!

```python
# Scatter + histogram
sns.jointplot(data=tips, x='total_bill', y='tip')
plt.show()

# KDE version
sns.jointplot(data=tips, x='total_bill', y='tip',
              kind='kde')
plt.show()

# Hex version
sns.jointplot(data=tips, x='total_bill', y='tip',
              kind='hex')
plt.show()

# With regression line
sns.jointplot(data=tips, x='total_bill', y='tip',
              kind='reg')
plt.show()
```

---

## 16. 📈 Regression Plot — regplot & lmplot

> Shows scatter plot with regression line!

```python
# Basic regression plot
sns.regplot(data=tips, x='total_bill', y='tip')
plt.show()

# lmplot with grouping
sns.lmplot(data=tips, x='total_bill', y='tip',
           hue='sex')
plt.show()

# In grid
sns.lmplot(data=tips, x='total_bill', y='tip',
           col='time', row='sex')
plt.show()
```

---

## 17. 🎨 Clustermap

> Heatmap with hierarchical clustering!

```python
iris = sns.load_dataset('iris')
iris_data = iris.drop('species', axis=1)

sns.clustermap(iris_data,
               cmap='viridis',
               standard_scale=1)
plt.show()
```

## 18. 🎨 Customization — Make Plots Beautiful!

```python
# Figure size
plt.figure(figsize=(10, 6))
sns.histplot(data=tips, x='total_bill', kde=True)
plt.show()

# Titles and labels
plt.figure(figsize=(10, 6))
sns.barplot(data=tips, x='day', y='total_bill')
plt.title('Average Bill by Day', fontsize=16)
plt.xlabel('Day of Week', fontsize=12)
plt.ylabel('Average Bill ($)', fontsize=12)
plt.show()

# Color palettes
sns.set_palette('husl')      # colorful
sns.set_palette('Blues')     # blue shades
sns.set_palette('Set2')      # soft colors
sns.set_palette('dark')      # dark colors

# Themes
sns.set_style('darkgrid')    # dark with grid
sns.set_style('whitegrid')   # white with grid
sns.set_style('white')       # plain white
sns.set_style('ticks')       # with tick marks

# Context - controls font sizes
sns.set_context('paper')     # small
sns.set_context('notebook')  # default
sns.set_context('talk')      # medium
sns.set_context('poster')    # large
```

## 19. 🎨 Multiple Plots Together

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1
sns.histplot(data=tips, x='total_bill',
             ax=axes[0,0], kde=True)
axes[0,0].set_title('Bill Distribution')

# Plot 2
sns.boxplot(data=tips, x='day', y='total_bill',
            ax=axes[0,1])
axes[0,1].set_title('Bill by Day')

# Plot 3
sns.scatterplot(data=tips, x='total_bill', y='tip',
                ax=axes[1,0], hue='sex')
axes[1,0].set_title('Bill vs Tip')

# Plot 4
sns.countplot(data=tips, x='day',
              ax=axes[1,1])
axes[1,1].set_title('Count by Day')

plt.tight_layout()
plt.show()
```

## 20. 🤖 Seaborn in ML Workflow

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load ML dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

# Step 1 — Check distributions
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for idx, col in enumerate(iris.feature_names):
    ax = axes[idx//2, idx%2]
    sns.histplot(data=df, x=col,
                 hue='target', ax=ax, kde=True)
plt.tight_layout()
plt.show()

# Step 2 — Check correlations
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True),
            annot=True, cmap='coolwarm')
plt.title('Feature Correlations')
plt.show()

# Step 3 — Pairplot for all features
sns.pairplot(df, hue='target')
plt.show()

# Step 4 — Check for outliers
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for idx, col in enumerate(iris.feature_names):
    sns.boxplot(data=df, y=col, ax=axes[idx])
plt.tight_layout()
plt.show()
```

## 📌 Quick Cheatsheet

| Plot | Use when | Code |
|---|---|---|
| histplot | Distribution of one variable | `sns.histplot(data, x='col')` |
| kdeplot | Smooth distribution | `sns.kdeplot(data, x='col')` |
| boxplot | Outliers and spread | `sns.boxplot(data, x, y)` |
| violinplot | Distribution shape | `sns.violinplot(data, x, y)` |
| barplot | Category averages | `sns.barplot(data, x, y)` |
| countplot | Count categories | `sns.countplot(data, x)` |
| scatterplot | Two numeric variables | `sns.scatterplot(data, x, y)` |
| lineplot | Trends over time | `sns.lineplot(data, x, y)` |
| heatmap | Correlation matrix | `sns.heatmap(corr)` |
| pairplot | All variable pairs | `sns.pairplot(data)` |
| jointplot | Relationship + distribution | `sns.jointplot(data, x, y)` |
| regplot | Scatter + regression | `sns.regplot(data, x, y)` |
